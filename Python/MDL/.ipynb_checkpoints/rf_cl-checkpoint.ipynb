# Random forest for imbalanced classification  

### Libs

```python
import pandas as pd
import numpy as np
import scikitplot 
import shap

from sklearn.datasets import make_classification
from imblearn.ensemble import BalancedRandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import average_precision_score, accuracy_score, classification_report,\
                            confusion_matrix, roc_curve, precision_recall_curve, f1_score,\
                            auc, precision_score, recall_score 

from matplotlib import pyplot as plt
from boruta import BorutaPy
from optuna import create_study
```

### Class 

```python
class imbal_RF(object):
    
    '''
    A class to generate classification data with imbalance for binary classification task.
    As a classifier, Balanced Random Forest will be used. To filter out redunat features,
    BorutaPy will be invoked. Then the reduced data set is split at random. Optuna will be
    used to select the most optimal hyperparameter values. Only with the training data shap
    values and shap plots as well as lift curve will be adopted to evaluate the most important
    features for the classification task and the abbility of the clasifier to predict the positive 
    class, given the imbalance in the data. In the end, classification metrics and reports are 
    at avail to summarize in greater detail the model performance with the testing data. 
    
    ...

    Attributes
    ----------
    data/target, data_filt, data_train/target_train, data_test/target_test, y_pred: np.array
        Arrays with the corresponding data sets as described in the docstring above.
    study: obj
        BorutyPy object
    ap_score, summary, precision, recall, thresholds, auc, f1, cm: obj
       Classification metrics
        
    
    Methods
    -------
    feature_opt():
        Performs feature selection with BorutaPy.
    rnd_split():
        Splits the filtered data at random.
    _obj_func(trial):
        Objective function for Optuna to evaluate the most optimal values of the hyperparameters.
    hypar_opt(_obj_func):
        Performs the hyperparamtertic optimization.
    train_predict_eval():
        Fitting the model after the optimizations, plotting the shap plots and the lift curves and
        computes all classification metrics.  
        
    '''
    
    __slots__ = ['data','target','data_filt', 'data_train', 'target_train', 'data_test', 'target_test', 'y_pred',
                 'study', 'ap_score', 'summary', 'precision', 'recall', 'thresholds', 'auc', 'f1', 'cm']
    
    def __init__(self):
        
        self.data, self.target = make_classification(n_samples = 10000, n_features = 100, n_informative = 75, 
                                                     n_classes = 2, weights = [0.60, 0.40], random_state = 24)
        print('Data successfully generated.')
        print('\nData (data) shape:', self.data.shape)
        print('\nTarget (target) shape:', self.target.shape)
    
    def feature_opt(self):
        
        print('\nFiltering with Boruta...\n')
        selector = BorutaPy(BalancedRandomForestClassifier(n_jobs = -1, class_weight = 'balanced',
                                                           sampling_strategy = 'majority', replacement = False),
                            n_estimators = 'auto', verbose = 2, random_state = 42).fit(self.data, self.target)

        self.data_filt = selector.transform(self.data)
        
        print('\nFiltered data (data_filt) shape:', self.data_filt.shape)
        
    def rnd_split(self):      
        
        print('\nSplitting at random...')
        self.data_train, self.data_test, self.target_train, self.target_test = train_test_split(self.data_filt, 
                                                                                                self.target, 
                                                                                                train_size = 0.9,
                                                                                                test_size =  0.1,
                                                                                                random_state = 42)
        print('\nData successfully splitted.')
        print('\nTrain data (data_train) shape:', self.data_train.shape)
        print('\nTrain target (target_train) shape:', self.target_train.shape)
        print('\nTest data (data_test) shape:', self.data_test.shape)
        print('\nTest target (target_test) shape:', self.target_test.shape)
        
    def _obj_func(self, trial):
        
        n_estimators = trial.suggest_int('n_estimators', 3, 21)
        min_samples_split = trial.suggest_int('min_samples_split', 2, 11)
        min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 11)
        
        clf = BalancedRandomForestClassifier(n_estimators = n_estimators, min_samples_split = min_samples_split,
                                             min_samples_leaf = min_samples_leaf, class_weight = 'balanced', 
                                             sampling_strategy = 'majority',replacement = False,
                                             random_state = 42, n_jobs = -1).fit(self.data_train, self.target_train)
         
        return average_precision_score(self.target_train, clf.predict_proba(self.data_train)[:,1]) 
    
    def hypar_opt(self, _obj_func):
        
        print('\nStarting the optimization of the hyperparameters with Optuna (Evaluation metric - ap_score)...\n')
        self.study = create_study(direction = 'maximize')
        self.study.optimize(_obj_func, n_trials = 10)
        print('\nOptimization completed.')
        print('\nHighest ap_score: %.2f%%' % (100*self.study.best_trial.value))    
        print('\nBest hyperparametrs:', self.study.best_trial.params)
        
    def train_predict_eval(self):
        
        print('\nInitializing Balanced Random Forest Classifier with the train data and the best hyperparameters...')
        
        clf = BalancedRandomForestClassifier(n_estimators = self.study.best_trial.params['n_estimators'], 
                                             min_samples_split = self.study.best_trial.params['min_samples_split'], 
                                             min_samples_leaf = self.study.best_trial.params['min_samples_leaf'],
                                             class_weight = 'balanced', sampling_strategy = 'majority',
                                             replacement = False, random_state = 42, 
                                             n_jobs = -1).fit(self.data_train,self.target_train)
        
        print('\nInitialization done.')
        print('\nCalculating the shap values and creating the shap plots...\n')
        
        explain = shap.TreeExplainer(clf, self.data_train)
        shap_vals = explain(self.data_train)
        shap.plots.bar(shap_vals[:,:,1], max_display = 10)
        shap.plots.beeswarm(shap_vals[:,:,1], plot_size = (12, 9), max_display = 10)
        
        print('\nPlotting the lift curve\n')
        scikitplot.metrics.plot_lift_curve(self.target_train, 
                                           clf.fit(self.data_train, self.target_train).predict_proba(self.data_train))
        
        
        
        print('\nMaking predictions with the testing data.')
        
        self.y_pred = clf.fit(self.data_train, self.target_train).predict_proba(self.data_test)
        self.ap_score = average_precision_score(self.target_test, np.round(self.y_pred[:,1]))
        print("\nAvererage precision score: %.2f%%" % (self.ap_score * 100.0))
        
        self.summary = classification_report(self.target_test, np.round(self.y_pred[:,1]))
        print('\nPrecision, recall, accuracy and f1-score:\n', self.summary)
        
        self.precision, self.recall, self.thresholds = precision_recall_curve(self.target_test, 
                                                                              np.round(self.y_pred[:,1]))
        
        plt.figure()
        plt.plot(self.recall, self.precision)
        plt.xlabel('Recall')
        plt.ylabel('Precision')
        plt.title('Precision - Recall Curve')
        plt.show()
        
        self.f1, self.auc = f1_score(self.target_test, np.round(self.y_pred[:,1])), auc(self.recall, self.precision)
        print('\nAUC:', f'{self.auc:.2}')
        
        self.cm = confusion_matrix(self.target_test, np.round(self.y_pred[:,1]))
        print('\nThe confusion matrix:\n', self.cm)
        
        
```

### Run

```python
model = imbal_RF()
model.feature_opt()
model.rnd_split()
model.hypar_opt(model._obj_func)
model.train_predict_eval()
```
