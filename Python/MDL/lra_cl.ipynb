```python
#GOAL: Use PCA algorithm to obtain low rank approximations of an input image.
```

```python
# Libs:
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

from scipy import linalg
from math import sqrt
```

```python
class lra(object):
    
    
    """
    A class to perform low rank approximations on an input image.
    
    ...

    Attributes
    ----------
    path_to_image : str
        Example: 'your_path/image.png'

    Methods
    -------
    input_image_load:
        loads the image from path_to_image
    input_image_show():
        Plots the original image
    lra_image_show(k):
        Plots the low rank approximated image, using PCA for k eigenvectors
    """   
    
    
    __slots__ = ['input_image_load']
    
    def __init__(self, path_to_image):
        self.input_image_load = mpimg.imread(path_to_image)
        print('Image size:', self.input_image_load.shape)
        
    def input_image_show(self):    
        plt.figure()
        plt.imshow(self.input_image_load)
            
    def init_lra(self, k):
        i, j, c = self.input_image_load.shape
        lra_image = np.zeros((i, j, c))

        for jj in range(c):
            M = np.mean(self.input_image_load[:,:,jj], axis=0)
            C = self.input_image_load[:,:,jj] - M
            u, l, ut = np.linalg.svd(C, full_matrices=False)
            sv = np.zeros((j, k))
            sv[:, :k] = ut[:k,:].T
            lra_image[:,:,jj] = C @ sv @ sv.T + M

        yield lra_image

    def lra_image_show(self, k):

        for lra_image in self.init_lra(k):
            plt.figure()
            plt.imshow(lra_image)
            plt.title(f'LRA Image, using PCA for k = {k}')
            plt.show()
    
```

```python
k_values = [1,10,100,1000]
```

```python
I = lra('ai2.jpg')
```

```python
I.input_image_load
```

```python
I.input_image_show()
```

```python
for k in k_values:
    I.lra_image_show(k)
```
